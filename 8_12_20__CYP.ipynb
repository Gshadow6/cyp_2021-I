```
#problema 3.1
sumpar = 0
sumimp = 0
cuepar = 0
propar = 0
i = 1
while (i <= 5):
  num = int(input("Introduce un numero entero "))
  if (num != 0):
    if ((-1)**num) > 0:
      print("Es par")
      sumpar = sumpar + num
      cuepar = cuepar + 1
    else:
      sumimp = sumimp + num
  i = i + 1
propar= sumpar/cuepar
print(f"El promedio de los pares es {propar}")
print(f"La suma de los impares es {sumimp}")
```
